```python
import os
import time

import numpy as np
import pandas as pd

from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
#Information we want to scrape
MainData = pd.DataFrame(columns = ['Summary','Date', 'Author', 'OverallRating', 'Pros', 'Cons', 'Helpful'])
```

```python
#Define a function that is reusable. We need to scrape many pages of many companies
def SCRAPER(url):
    global MainData
    
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = Request(url,headers=hdr)
    page = urlopen(req)
    soup = BeautifulSoup(page, "html.parser")
    
    #Get the summary
    Summary = soup.find_all('a', attrs = {'class':'reviewLink'})
    SummaryClean = []
    for x in Summary:
        SummaryClean.append(x.text)
    
    #Get Postdate
    PostDate = soup.find_all('span', attrs = {'class':'authorInfo'})
    PostDateClean = []
    for x in PostDate:
        PostDateClean.append(x.text)
        
    #Get the Author status
    AuthorStatus = soup.find_all('span', attrs = {'class': 'pt-xsm pt-md-0 css-1qxtz39 eg4psks0'})
    AuthorStatusClean = []
    for x in AuthorStatus:
        AuthorStatusClean.append(x.text)
        
    #Get overall rating
    ValueTitle = soup.find_all('span', attrs = {'class':'ratingNumber'})
    ValueTitleClean = []
    for x in ValueTitle:
        ValueTitleClean.append(x.text)
    
    #Get Pros
    Pros = soup.find_all('span', attrs = {'data-test':'pros'})
    ProsClean = []
    for x in Pros:
        ProsClean.append(x.text)  
    
    #Get Cons
    Cons = soup.find_all('span', attrs = {'data-test':'cons'})
    ConsClean = []
    for x in Cons:
        ConsClean.append(x.text)    
    
    #Get How many people found review helpful
    Helpful = soup.find_all('div', attrs = {'class':'common__EiReviewDetailsStyle__socialHelpfulcontainer pt-std'})
    HelpfulClean = []
    for x in Helpful:
        HelpfulClean.append(x.text) 
    
    DATA = pd.DataFrame(list(zip(SummaryClean, PostDateClean,  AuthorStatusClean, ValueTitleClean, ProsClean, ConsClean, HelpfulClean)), columns = ['Summary','Date', 'Author', 'OverallRating', 'Pros', 'Cons', 'Helpful'])
    MainData=MainData.append(DATA,)
    #Wait a little
    time.sleep(30)
```

```python
for x in range(1,71):
    url="https://www.glassdoor.com/Reviews/Zebra-Technologies-Reviews-E2089_P"+str(x)+".htm?sort.sortType=RD&sort.ascending=false"
    SCRAPER(url)
#write files
MainData=MainData.reset_index(drop=True)
MainData.to_csv('ZebraTechnologies_Glassdoor_Data.csv', mode = 'w')
```
